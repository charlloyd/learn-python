```python
### NUMBER OPERATIONS

print(2+3)
print(3-2)
print(3*4)

### INTEGERS

a=2
b=3
c="4" #this is a number encoded as a string variable

print() #prints a blank line
print(a+b)
print(b-a)
print(b*int(c)) #don't forget to convert c from a string to an integer before using it in integer operations
print(int(c)/a)

### FLOATS (non-integer numbers, i.e. numbers with decimals)

print()
print(4/2) #the result of division is always returned as a float even if it is an integer
print(9/2) 
print(9%2) #prints the remainder of 9 divided by 2
print(int(9/2)) #converting a float to an integer always rounds down
print(float(2)) #converts an integer to a float
```

```python
### STRINGS

d="apple"
e="banana"
f="meat"+"loaf" #this concatenates two strings (meat and loaf) to become a single string (meatloaf) saved as variable d

print(d)
print(e)
print(f)

### string anatomy
print()
print(f[0]) #prints the character in the 0 position of string c (i.e. the first letter)
print(f[:4]) #prints from the beginning until (not including) the character in the 4 position
print(f[4:]) #prints from (including) the character in the 4 position until the end
print(f[-1]) #prints the character in the last position
print(f[2:6]) #prints 4 characters from positions 3 to 6 (inclusive)
print(len(f)) #prints the length (i.e. number of characters) of the string

### printing strings
print()
print("number of characters: "+str(len(f))) #prints the concatenation of two strings (first converts the integer len(f) to a string))
print("number of characters: %s" % len(f)) #prints a string + varaible using %-format
print("There are %s characters in %s." % (len(f),f)) #prints a string and two variables using %-format
print("The word \"%s\" has %s characters." % (f,len(f))) #prints using %-format and "escapes" the quote characters using the backslash / (otherwise they will be read as code, not characters to print).
```

```python
### LISTS

g=["stuff","things"]
h=[a,b,c]
i=[d,e,f]

print(g)
print(h)
print(i)

### list anatomy
print()
print(g[0]) #prints the item in position 0 in list g
print(int(h[2])) #prints string in position 2 in list h converted to an integer
print(i[2][4:]) #prints the characters from position 4 to the end of the string in position 2 in list i

### concatenate & extend
print()
j = h+i #creates a new list in memory saved as j that is the combo of h and i (h and i remain unchanged) 
print(j)
j.extend(['banana', 'banana']) #adds a list to h (modifies h)
print(j)

#### append & insert
print()
j.append("swimming pool") #appends a single string to the end of list j
print(j)
j.append(g) #appends the list g to j
print(j)
j.insert(4,'banana')
print(j)

### count
print()
print("Number of bananas in the list: %s" % j.count('banana'))

### pop
print()
print(j.pop()) #pops (deletes) the last item from k
print(j)
k=j.pop() #pops (deletes) the last item from k AND saves it as j
print(k)
print(j)
print(j.pop(2)) #pops (deletes) the item in position 2
print(j.pop(-6)) #pops (deletes) the item in position that is 6th from the end of the list
print(j)

### remove
print(j)
j.remove('banana') #removes the first instance of banana

### list comprehension - a fancy and convenient way to do stuff with lists
print()
print([l for l in j]) #prints all items in list j exactly as they are
print([l[0] for l in j[2:]]) #prints the first character of each string in list j from positions 3 to the end (inclusive)

#NOTE: lists can be tricky to modify because of the way they are saved in memory, so always verify your changes
```

```python
### FOR loops

for m in [1,2,3]:
    print(m) # prints all the integers in the given list
    
print()
for n in i: 
    print(n) #prints all the strings in list i

print()
for o in i[2]:
    print(o) #prints all the letters in the string in the number 2 position of list i

print()
for p in range(5):
    print(p) #prints the numbers from 0 to 4

print()
for q in range(len(i)):
    print(q) #prints the numbers from 0 to the length of the list i

### WHILE loops
print()
r=1
while r<5:
    r+=1
    print(r)
    
print()
s=0
while s<len(i):
    print(i[s])
    s+=1
```

```python
### CONDITIONALS

t=5

if t>=4:
    print("True")
else:
    print("False")
    
if len("meatloaf")==8:
    t+=5
print(t)

print()
while t>0:
    if t%2==0:
        print(t) #prints even numbers backwards from 10
    t-=1
```

```python
### FUNCTIONS

u = [1,2,3]

def square(v):
    return v*v

def exponent (w,x):
    return w**x

print(square(5))
print(exponent(5,2))
print(exponent(5,3))

print()
for w in u:
    print("Calculating for w=%s" % str(w))
    print(square(w))
    print(exponent(w,3))
    print(exponent(w,4))
    print(exponent(w,5))
```

```python
### READING DATA INTO PYTHON

import csv #you will need to use the "CSV" package for this

urls_filepath = '/Users/charlottejolloyd/Dropbox/Projects/scraping-workshops/sandbox-2/urls.txt'
data_filepath = '/Users/charlottejolloyd/Dropbox/Projects/scraping-workshops/sandbox-2/data.csv'

### read a csv or txt file into a single list

urls = [line for line in open(urls_filepath,'r')] #reads all lines from the urls txt into a list (NOTE: /n is a new line character)
print(urls[:5])

print()
urls = [line.strip() for line in open(urls_filepath,'r')] #reads all lines from the urls txt into a list AND strips off line endings at the same time
print(urls[:5])

print()
data = [line.strip() for line in open(data_filepath,'r')] #reads all lines from the data csv (NOTE: each line is one string including commas)
print(data[:5])

#read a csv file into a list of lists
print()
with open(data_filepath, 'r') as f:
    data = list(list(r) for r in csv.reader(f, delimiter=','))
print(data[:5])

for t in data[:2]:
    print()
    print(t)
    for u in t:
        print(u)
```

```python
### WRITING DATA OUT OF PYTHON

urls2_filepath = '/Users/charlottejolloyd/Dropbox/Projects/scraping-workshops/sandbox-2/urls2.txt'
data2_filepath = '/Users/charlottejolloyd/Dropbox/Projects/scraping-workshops/sandbox-2/data2.csv'

urls.append("this is a fake URL")
print(urls[-5:])
print()
data.insert(0,['Month','Integer','Year','Letter','Fruit'])
print(data[:5])

#write all elements of a list to an individual line in to a BLANK text or csv file
f=open(urls2_filepath,'w')
f.writelines(["%s\n" %v for v in urls])
f.close()

#write a list of lists to a BLANK csv file
import csv
with open(data2_filepath, 'w') as f:
    writer = csv.writer(f)
    writer.writerows(data)
    f.close()
```

```python
### APPENDING DATA TO EXISTING FILES IN PYTHON 

#add a single line to the EXISTING text or csv file 
f=open(urls2_filepath,'a')
f.writelines(['adding another line'])
f.close()
 
#add multiple lines to an EXISTING text or csv file
with open(data2_filepath, 'a') as f:
    writer = csv.writer(f)
    writer.writerows(data[1:]) #adds the entire data list of lists except for the first line to the file again 
    f.close()
```

```python
### GRAPHS

import matplotlib.pyplot as plt #one of many libraries for graphing, not necessarily the best

x = [1,2,3,4,5]
y = [3,7,1,5,8]
y2 = [6,3,2,9,4]

### LINE GRAPH

plt.plot(x, y, '-r', label='x')
plt.plot(x, y2, '-b', label='y')

plt.xlabel('x')
plt.ylabel('y')
plt.title('X vs Y')
plt.legend(loc=2)
plt.show()

### BAR GRAPH

plt.bar(x, y, width=.5, color='b', label="y")

plt.xlabel('x')
plt.ylabel('y')
plt.title('X')
plt.legend(loc=2)
plt.show()

### STACKED BAR GRAPH

plt.bar(x, y, width=.5, color='b', label="y")
plt.bar(x, y2, width=.5, color='r', bottom=y, label="y2")

plt.xlabel('x')
plt.ylabel('y')
plt.title('X vs Y')
plt.legend(loc=2)
plt.show()
```
